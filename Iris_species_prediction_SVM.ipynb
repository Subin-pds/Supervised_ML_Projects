```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
iris.feature_names
```

```python
df = pd.DataFrame(iris.data,columns=iris.feature_names)
df.head()
```

```python
df['target'] = iris.target
df.head()
```

```python
df[df['target']==1].head()
```

```python
df[df['target']==2].head()
```

```python
df['flower_name'] = df.target.apply(lambda x: iris.target_names[x])
df.head()
```

```python
df0 = df[df['target']==0]
df1 = df[df['target']==1]
df2 = df[df['target']==2]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(5,4))
s1 = plt.scatter(df0['sepal length (cm)'], df0['sepal width (cm)'], color='green', marker='*',label=str(iris.target_names[0]))
s2 = plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'], color = 'red', marker = '+',label=str(iris.target_names[1]))
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.legend(handles=[s1,s2])
plt.show()
```

```python
plt.figure(figsize=(5,4))

s1 = plt.scatter(df0['petal length (cm)'], df0['petal width (cm)'], color='green', marker='*', label=str(iris.target_names[0]))
s2 = plt.scatter(df1['petal length (cm)'], df1['petal width (cm)'], color = 'red', marker = '+',label=str(iris.target_names[1]))
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')
plt.legend(handles=[s1,s2])
plt.show()
```

```python
X = df.drop(['target','flower_name'], axis='columns')
y = df.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=.2, random_state=0)
```

```python
from sklearn.svm import SVC
model = SVC()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_train,y_train)
```

```python
# Tune Parameters : Regularization (C)
model_c = SVC(C=100)
model_c.fit(X_train,y_train)
model_c.score(X_test,y_test)
```

```python
# Tune Parameters : Gamma
model_g = SVC(gamma=100)
model_g.fit(X_train,y_train)
model_g.score(X_test,y_test)
```

```python
# Tune Parameters : Kernal
model_linear_kernal = SVC(kernel='linear')
model_linear_kernal.fit(X_train,y_train)
model_linear_kernal.score(X_test,y_test)
```
