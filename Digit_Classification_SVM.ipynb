```python
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
digits.target
```

```python
dir(digits)
```

```python
digits.target_names
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(digits.data, digits.target)
df.head()
```

```python
df['target'] = digits.target
```

```python
df['target'].value_counts()
```

```python
X = df.drop(['target'], axis='columns')
y = df.target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state = 3)
X_test.shape
```

```python
from sklearn.svm import SVC
model = SVC()
```

```python
model.fit(X_train,y_train)
```

```python
print("Training Accuracy: ", model.score(X_train, y_train))
print("Test Accuracy: ", model.score(X_test, y_test))
```

```python
# Parameter Tuning : Kernal 
model_linear_kernal = SVC(kernel='linear')
model_linear_kernal.fit(X_train,y_train)
print("Training Accuracy: ", model_linear_kernal.score(X_train, y_train))
print("Test Accuracy: ", model_linear_kernal.score(X_test, y_test))
```

```python
# Parameter Tuning : gamma
model_gamma = SVC(gamma=100)
model_gamma.fit(X_train,y_train)
print("Training Accuracy: ", model_gamma.score(X_train, y_train))
print("Test Accuracy: ", model_gamma.score(X_test, y_test))
```

```python
# Parameter Tuning : Regularization (C)
model_c = SVC(C=100)
model_c.fit(X_train,y_train)
print("Training Accuracy: ", model_c.score(X_train, y_train))
print("Test Accuracy: ", model_c.score(X_test, y_test))
```
